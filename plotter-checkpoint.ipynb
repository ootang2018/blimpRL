# Plotting Tools

```python
import os
import csv

import numpy as np
from scipy.io import loadmat, savemat
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import seaborn as sns; sns.set()
```

### Reward

```python
log_dir = "/home/rtallamraju/blimp_ws/exp_log/pets/HOVER/4act/"  # Directory specified in script, not including date+time
time_stamp = "rate2_hor10_iter1000_NN3by250_continue1"
sleep_rate=2

min_num_trials = 250# Plots up to this many trials


returns = []

data = loadmat(os.path.join(log_dir, time_stamp, "logs.mat"))
if data["returns"].shape[1] >= min_num_trials:
    returns.append(data["returns"][0][:min_num_trials])
        
returns = np.array(returns)/30/sleep_rate*100
# print(returns)
# returns = np.maximum.accumulate(returns, axis=-1)
mean = np.mean(returns, axis=0)

# Plot result
plt.figure()
axes = plt.gca()
axes.set_xlim([0,min_num_trials])
axes.set_ylim([-100,0])
plt.plot(np.arange(1, min_num_trials + 1), mean)
plt.title("Performance")
plt.xlabel("Iteration number")
plt.ylabel("Return")
plt.show()
```

### ALT

```python
log_dir = "/home/rtallamraju/catkin_ws_py3/src/rl_log/pets_mpc/ALT/3act/"  # Directory specified in script, not including date+time
time_stamp = "exp_hor20_rate10"

use_mpc = True
sleep_rate= 10
min_num_trials = 250# Plots up to this many trials

#*--- Start ---*#
time_step = 30*sleep_rate

data = loadmat(os.path.join(log_dir, time_stamp, "logs.mat"))

z = np.array(data['observations'][:,:,8])       
act =  np.array(data['observations'][:,:,1:9])
r = data["returns"][0][:min_num_trials]


z = np.abs(z)
if use_mpc:
    z = -np.tanh(0.5*z)
    r_mpc = np.array(r)/30/sleep_rate*100
else:
    z = -np.tanh(0.1*z)

act = np.linalg.norm(act, axis=1)
act =  -np.tanh(0.2*act)    

r_z = np.mean(z, axis=1)*100
r_a = np.mean(act, axis=1)*100
reward = 0.9*r_z + 0.1*r_a


# # Plot result
# plt.figure()
axes = plt.gca()
axes.set_xlim([0,min_num_trials])
axes.set_ylim([-100,-0])

if use_mpc:
    plt.plot(np.arange(1, min_num_trials+1 ), r_mpc, label='reward')
    plt.plot(np.arange(0, min_num_trials+1 ), reward, label='mpc_cost')
else:
    plt.plot(np.arange(0, min_num_trials+1 ), reward, label='reward')

plt.plot(np.arange(0, min_num_trials+1 ), 0.1*r_a, label='action_cost')
plt.title("Performance")
plt.xlabel("Iteration number")
plt.ylabel("Return")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

### TAKEOFF

```python
log_dir = "/home/rtallamraju/blimp_ws/exp_log/pets/HOVER/4act/"  # Directory specified in script, not including date+time
time_stamp = "rate2_hor10_iter1000_NN3by250_continue1"

use_mpc = False
sleep_rate= 2
min_num_trials = 1000# Plots up to this many trials

#*--- Start ---*#
time_step = 30*sleep_rate

data = loadmat(os.path.join(log_dir, time_stamp, "logs.mat"))

x = np.array(data['observations'][:,:,6:9]) 
ang = np.array(data['observations'][:,:,0:3])    
act =  np.array(data['actions'][:,:,1:9])
r = data["returns"][0][:min_num_trials]

x = np.linalg.norm(x, axis=2)
if use_mpc:
    x = -np.tanh(0.5*x)
    r_mpc = np.array(r)/30/sleep_rate*100
else:
    x = -np.tanh(0.1*x)
    
ang = np.mean(np.abs(ang), axis=1)
ang = -np.tanh(1*ang)

act = np.linalg.norm(act, axis=1)
act =  -np.tanh(0.2*act)    

r_x = np.mean(x, axis=1)*100
r_ang = np.mean(ang, axis=1)*100
r_act = np.mean(act, axis=1)*100
reward = 0.8*r_x + 0.1*r_ang + 0.1*r_act


# # Plot result
# plt.figure()
axes = plt.gca()
axes.set_xlim([0,min_num_trials])
axes.set_ylim([-100,-0])

if use_mpc:
    plt.plot(np.arange(1, min_num_trials+1 ), r_mpc, label='reward')
    plt.plot(np.arange(0, min_num_trials+1 ), reward, label='mpc_cost')
else:
    plt.plot(np.arange(0, min_num_trials+1 ), reward, label='reward')

plt.plot(np.arange(0, min_num_trials+1 ), r_x, label='dist_cost')
plt.plot(np.arange(0, min_num_trials+1 ), r_ang, label='ang_cost')
plt.plot(np.arange(0, min_num_trials+1 ), r_act, label='act_cost')
plt.title("Performance")
plt.xlabel("Iteration number")
plt.ylabel("Return")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
